Up until now, we have been the Client (sending requests). Now, we are going to build the Server (receiving requests).

FastAPI is the industry standard for AI backends because it natively supports the async concepts we just learned, and it pairs perfectly with Pydantic, which acts like a strict bouncer for your data.

FYI: We can't run the code in a notebook because it will freeze the notebook, check fastapi_and_pydantic.py!

Core packages:
1. FastAPI: The web framework itself.
2. Uvicorn: The lightning-fast asynchronous server that actually runs your FastAPI code (it acts as the middleman between the internet and your Python code).
3. Pydantic: The data validation library.

```python
pip install fastapi "uvicorn[standard]" pydantic
```

to run the actual server you need to do the following: 
1. uvicorn fastapi_and_pydantic:app --reload
2. go to the web address it says it is running on 

Breaking down the command:
* uvicorn: The server program.
* fastapi_and_pydantic: The name of your Python file (notice there is no .py).
* :: The separator telling Uvicorn to look inside that file.
* app: The name of the variable where you initialized your FastAPI server (app = FastAPI(...)).
* --reload: The flag that restarts the server whenever you save the file.

FastAPI automatically builds a testing interface for you.
* Open your web browser and go to: http://localhost:8000/docs
* You will see a "Swagger UI" dashboard.
* Click on the green POST /generate bar, click "Try it out", modify the JSON payload it gives you, and hit "Execute".
