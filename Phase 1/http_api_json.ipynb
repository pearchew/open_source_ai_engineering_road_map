GET vs POST requests
* GET request as reading a public notice board. You are asking the server to hand you a specific piece of information.
* Think of a POST request as mailing a sealed package. You are sending a bundle of data to the server for it to process, save, or act upon.

If you want to ask an LLM "What is the capital of France?", you could theoretically send it as a GET request like api.openai.com/ask?prompt=What+is....

However, we use POST for three main reasons:
1. Size: LLM prompts often include thousands of tokens of context, system instructions, and conversation history. That easily breaks the URL size limit of a GET request.
2. Complexity: LLM requests require complex, nested settings (temperature, max tokens, stop sequences). Sending this as a structured JSON body in a POST request is much cleaner than turning it into a massive, messy URL string.
3. Privacy: You don't want your users' private prompts (which might contain sensitive data) saved in plain text in network routing logs, which happens with URLs.

```python
# GET request is used to retrieve data.
import requests
import json

# 1. Make the GET request to a free test API
response = requests.get("https://jsonplaceholder.typicode.com/posts/1")

# 2. Check the status code (200 means OK)
print(f"Status Code: {response.status_code}")

# 3. Extract the JSON payload from the response
data = response.json() 

# 4. Print the parsed Python dictionary
print("Response Data:")
print(json.dumps(data, indent=2)) # json.dumps formats it nicely for reading
```

```python
# post requests
    # When you talk to an LLM, you are sending a lot of data: your prompt, system instructions, temperature, etc. Because of this, you use a POST request. 
    # You also have to tell the server what kind of data you are sending (JSON) and often provide an API key for authentication.
import requests

url = "https://jsonplaceholder.typicode.com/posts" # A test endpoint that accepts POST

# 1. Define your headers (Authentication and Content-Type)
headers = {
    "Content-Type": "application/json",
    "Authorization": "Bearer YOUR_FAKE_API_KEY"
}

# 2. Define your JSON payload (a Python dictionary)
payload = {
    "model": "gpt-fake-4",
    "messages": [{"role": "user", "content": "Hello, AI!"}],
    "temperature": 0.7
}

# 3. Make the POST request
response = requests.post(url, headers=headers, json=payload)

print(f"Status Code: {response.status_code}") # Expecting 201 (Created)
print("Server Response:")
print(response.json())
```

httpx

The requests library is an industry classic, but it is strictly synchronous (it blocks your code from doing anything else while waiting for the server). Since we will need asynchronous code for LLMs later, we will transition to httpx. It works almost identically to requests for synchronous code!

```python
import httpx

# Making the exact same GET request, but with httpx
with httpx.Client() as client:
    response = client.get("https://jsonplaceholder.typicode.com/posts/1")
    print(f"httpx Status: {response.status_code}")
    print(response.json()['title'])
```
