Step 2 -> containerising vector database
* Spin up a vector database (ChromaDB) to store your RAG embeddings, and learn how to make that data persist even if the container crashes or is deleted.

By design, containers are ephemeral. If you delete a container, everything inside it—including your database records—vanishes. To solve this, we use a Volume, which mounts a folder from your physical hard drive directly into the container.

The following needs to be done in the terminal:

docker run -d \
  --name chromadb \
  -p 8001:8000 \
  -v chroma-data:/chroma/chroma \
  -e IS_PERSISTENT=TRUE \
  chromadb/chroma:latest

* -d: Run in detached mode (background).
* --name chromadb: Gives the container a recognizable name instead of a random string of characters.
* -p 8001:8000: Maps port 8001 on your computer to port 8000 inside the Chroma container.
    * -p flag in Docker, the format is always Host_Port : Container_Port
    * The Left Side (8001) -> This is the port on your physical computer (your Mac) : The Right Side (8000) -> This is the port inside the isolated Docker container.
    * Because we previously set up docker fastapi container on desktop port 8000, we are linking the new api port to 8001 -> but actually they all just route throguh the computer
* -v chroma-data:/chroma/chroma: Creates a persistent Docker volume named chroma-data and attaches it to the directory where Chroma saves its database files.
* -e IS_PERSISTENT=TRUE: Sets an environment variable inside the container telling Chroma to save data to the disk.
* chromadb/chroma:latest: The exact image to pull from Docker Hub.

verify it is running -> docker ps
* docker ps, you are asking Docker to print a list of all the containers that are currently active and running in the background.
