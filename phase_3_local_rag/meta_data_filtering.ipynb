Vector embeddings are incredible at finding semantic meaning, but they are absolutely terrible at understanding hard constraints like dates, authors, or specific categories.

* Vector Search: "Find me books about how to fix a leaky pipe." (The librarian finds books with similar meaning).
* The Problem: The librarian hands you a plumbing manual from 1985. It's semantically a perfect match, but the tools it recommends don't exist anymore!
* Metadata Filtering: "Find me books about how to fix a leaky pipe, but only if they were published after 2022 and written by the Maintenance Department."

**As such, the method to help with this is to attach metadata during chunking. Then during the similarity search you can attach a filter!**

```python
# attaching metadata to chunks
chunk.metadata = {
    "source": "employee_handbook.pdf",
    "department": "HR",
    "year": 2024
}
```
