```python
import ollama

# We pass a list of messages. Right now, it's just one message from the 'user'.
response = ollama.chat(model='llama3.1', messages=[
    {
        'role': 'user',
        'content': 'Explain the concept of an LLM in one short sentence.',
    },
])

# The response is a dictionary. We extract just the text content of the message.
print(response['message']['content'])
```

```python
import ollama

# Notice we now have TWO messages in our list: the system instructions and the user query.
messages = [
    {
        'role': 'system',
        'content': 'You love shakeaspear, so all your responses read like a poem.'
    },
    {
        'role': 'user',
        'content': 'What is the best way to learn how to code?'
    }
]

response = ollama.chat(model='llama3.1', messages=messages)
print(response['message']['content'])
```
