What Actually is a Token? A token is not exactly a word. LLMs read text by breaking it down into chunks of characters.
* A short, common word like apple is usually 1 token.
* A longer or less common word like hamburger might be split into 3 tokens: ham, bur, ger.
* Punctuation marks and spaces also count.
* A good rule of thumb for English text is: 1 token $\approx$ 0.75 words (or 100 tokens $\approx$ 75 words).

Some initial simple things you can do: reduce conversation filler, use structure like xml tags, use few shot examples (Instead of writing 20 rules on how the model should behave, just show it 2 or 3 examples of the input and the expected output - allowing it to learn patterns)

```python
import tiktoken
```

```python
# 1. Load the standard tokenizer (used by GPT-4 and acts as a great baseline)
encoding = tiktoken.get_encoding("cl100k_base")

# 2. Define a classic "fluffy" prompt
fluffy_prompt = """
Hello there! I hope you are having an absolutely wonderful day today. 
I was wondering if you could possibly do me a huge favor. Could you please 
read through the following data and extract the names of the customers for me? 
I would really, really appreciate it. Thanks in advance!
Data: John Doe bought a car. Jane Smith bought a boat.
"""

# 3. Define the optimized, token-efficient version
optimized_prompt = """
Extract customer names.
Data: John Doe bought a car. Jane Smith bought a boat.
"""

# 4. Count the tokens
fluffy_tokens = encoding.encode(fluffy_prompt)
optimized_tokens = encoding.encode(optimized_prompt)

# 5. Calculate the difference
fluffy_count = len(fluffy_tokens)
optimized_count = len(optimized_tokens)
saved_tokens = fluffy_count - optimized_count
percentage_saved = round((saved_tokens / fluffy_count) * 100)

print("📊 Token Efficiency Report\n" + "-"*30)
print(f"❌ Fluffy Prompt:    {fluffy_count} tokens")
print(f"✅ Optimized Prompt: {optimized_count} tokens")
print("-" * 30)
print(f"💡 You saved {saved_tokens} tokens ({percentage_saved}% reduction)!")

# Bonus: Let's see EXACTLY how it sliced up the optimized prompt
print("\n🔍 How the AI sees the fluffy prompt:")
for token in fluffy_tokens:
    # We decode each individual token back to text to see the chunks
    print(f"[{encoding.decode([token])}]", end=" ")
print("\n🔍 How the AI sees the optimized prompt:")
for token in optimized_tokens:
    # We decode each individual token back to text to see the chunks
    print(f"[{encoding.decode([token])}]", end=" ")
```
